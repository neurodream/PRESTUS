```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
t = np.linspace(0, 10, 1000)  # Time vector
f = 1  # Frequency in Hz
y = np.sin(2 * np.pi * f * t)  # Sine wave

# Plot
plt.plot(t, y)
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Oscillation')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
t = np.linspace(0, 10, 1000)  # Time vector
f = 10  # Frequency in Hz
y = np.sin(2 * np.pi * f * t)  # Sine wave

# Boxcar function
boxcar = np.zeros_like(t)
boxcar[(t >= 2) & (t <= 3)] = 1  # First boxcar from 2 to 3 seconds
boxcar[(t >= 8) & (t <= 9)] = 1  # Second boxcar from 8 to 9 seconds

# Modulate the sine wave with the boxcar function
modulated_signal = y * boxcar

# Plot
plt.figure(figsize=(12,2))
plt.plot(t, modulated_signal, color="black")
# plt.xlabel('Time (s)')
# plt.ylabel('Amplitude')
# plt.title('Modulated Oscillation with Boxcar Function')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
t = np.linspace(0, 12.5, 1000)  # Time vector
f = 15  # Frequency in Hz (5 Hz)
y = np.sin(2 * np.pi * f * t)  # Sine wave

# Boxcar with ramping
boxcar = np.zeros_like(t)
ramp_time = 0.2  # Ramp duration in seconds

# Parameters for boxcar width and center positions
boxcar_width = 3  # Width of the boxcar in seconds
center_1 = 2.5  # Center of the first boxcar
center_2 = 8.5  # Center of the second boxcar

# First boxcar with ramps
start_1 = center_1 - boxcar_width / 2
end_1 = center_1 + boxcar_width / 2
boxcar[(t >= start_1 + ramp_time) & (t <= end_1 - ramp_time)] = 1  # Flat part
ramp_up_1 = (t >= start_1) & (t < start_1 + ramp_time)
ramp_down_1 = (t > end_1 - ramp_time) & (t <= end_1)
boxcar[ramp_up_1] = (t[ramp_up_1] - start_1) / ramp_time
boxcar[ramp_down_1] = 1 - (t[ramp_down_1] - (end_1 - ramp_time)) / ramp_time

# Second boxcar with ramps
start_2 = center_2 - boxcar_width / 2
end_2 = center_2 + boxcar_width / 2
boxcar[(t >= start_2 + ramp_time) & (t <= end_2 - ramp_time)] = 1  # Flat part
ramp_up_2 = (t >= start_2) & (t < start_2 + ramp_time)
ramp_down_2 = (t > end_2 - ramp_time) & (t <= end_2)
boxcar[ramp_up_2] = (t[ramp_up_2] - start_2) / ramp_time
boxcar[ramp_down_2] = 1 - (t[ramp_down_2] - (end_2 - ramp_time)) / ramp_time

# Modulate the sine wave with the ramped boxcar function
modulated_signal = y * boxcar

# Plot
plt.figure(figsize=(12,2))
plt.plot(t, modulated_signal, color="black")
plt.show()
```
