```python
import os
from pathlib import Path
from IPython.display import Image

simdata_path = Path("../../data/sims")
[f for f in os.listdir(simdata_path) if "_FW_" in f]
```

```python
def show_image(fname, literal=None, transd=None, exp=None, units=None, verbose=False, width=600):
    
    # requires "FW": only free water
    found_sims = [f for f in os.listdir(simdata_path) if "FW" in f.split("_")]
    if literal:
        found_sims = [f for f in found_sims if literal in f.split("_")]
    if transd:
        found_sims = [f for f in found_sims if transd in f.split("_")]
    if exp:
        found_sims = [f for f in found_sims if f"exp{exp}" in f.split("_")]
    if units:
        found_sims = [f for f in found_sims if f"{units}units" in f.split("_")]

    if verbose:
        for sim in found_sims:
            print(sim)
    
    # if multiple found, only show one (first)
    print(found_sims[0].split("_FW_")[1])
    display(Image(filename=simdata_path/found_sims[0]/f"{fname}.png", width=600))
```

## 1. transducer profile  
assuming it just depends on transducer profile, sanity check: (passed)

```python
show_image("transducer_profile", transd="IMASONIC100")
show_image("transducer_profile", transd="CTX")
```
